```python
image = cv.imread("tag.png")
plt.imshow(image) # BGR vs RGB, unimportant once we greyscale
```

## Thresholding

```python
blurred = cv.GaussianBlur(image, (3,3), 0) 
threshold = cv.threshold(blurred, 100, 250, cv.THRESH_BINARY)
edges = cv.Canny(blurred, 200, 150)
plt.imshow(edges)
```
